### Calculate ISA (International Standard Atmosphere)

```python
from math import log, exp, sqrt 

def isa(altitude):
   # return -> pressure, density, temperature -> depending on provided altitude.
   # Units are SI (altitude [m], pressure [Pa], densisty [kg/m^3], temperature [K])
    
    if altitude<=11000: 
        # troposphere
        pressure = 101325 * (1 - 2.25569E-5 * altitude)**5.25616
        density = 1.225 * (1 - 2.25569E-5 * altitude)**4.257
        temperature = 288.14 - 0.00649 * altitude     
        
    elif altitude<=20000:  
        # stratosphere
        pressure = 0.223356 * 101325 * exp(-0.000157688 * (altitude - 11000))  
        density = 0.296838*1.225*exp(-0.000157688*(altitude - 11000))
        temperature = 216.66  
        
    else:
        raise ValueError('altitude out of range [0-20000m]')
        
    return (pressure, density, temperature)


altitudes = range(0, 20100, 100) # altitudes from 0 to 20000m every 1000m

# calculate (pressure, density, temperature) for the given range and put them in a list
values = [isa(alt) for alt in altitudes]  # all altitudes

# pressures, densities, temperatures into three lists

pressures = [data[0] for data in values]  
densities = [data[1] for data in values] 
temperatures = [data[2] for data in values]

print(values) 
```
